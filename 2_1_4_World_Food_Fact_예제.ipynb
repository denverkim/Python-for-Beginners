# World Food Facts 연습문제

```python
# World Food Facts 파일 열기
import pandas as pd

food = pd.read_csv('/content/en.openfoodfacts.org.products.tsv', delimiter='\t')

# 처음 5줄
food.head()
```

```python
# 관측치의 갯수
print('관측치의 갯수: ', len(food))
print(food.info())
print(food.shape[0])

# 컬럼의 수
print('컬럼의 수: ', len(food.columns))
print(food.shape[1])

# 모든 컬럼의 이름
print('모든 컬럼의 이름: ', food.columns)

# 컬럼의 이름
print('컬럼의 이름: ', food.columns)
```

```python
# 105번째 컬럼 이름
print('105번째 컬럼 이름: ', food.columns[104])

# 105번째 컬럼 데이터 타입
print('105번째 컬럼 데이터 타입: ', food.dtypes[104])

# 인덱스
print('인덱스: ', food.index)

# 19번째 관측치 상품이름
print('19번째 관측치 상품이름: ', food.product_name[18])
food.loc[18,'product_name']
```

# Chipotle 연습문제

```python
import pandas as pd
chipo = pd.read_excel('/content/chipotle.xlsx')

# 상위 10개 데이터
chipo.head(10)
```

```python
print('관측치 갯수: ', chipo.shape[0])
print('컬럼의 수: ', chipo.shape[1])
print('모든 컬럼의 이름: ', chipo.columns)
print('인덱스: ', chipo.index)
```

```python
print('전체 주문 아이템의 수: ', chipo.quantity.sum())
print('전체 수익: ', (chipo.quantity * chipo.item_price).sum())
```

```python
# 가장 많이 주문된 아이템
chipo.groupby('item_name').count().order_id.sort_values(ascending=False).index[0]

# 가장 많이 주문된 아이템 갯수
chipo.groupby('item_name').count()['order_id'].sort_values(ascending=False)[0]

# 빈도 테이블
chipo.item_name.value_counts()
```

```python
# 가장 많이 선택한 choice_description은? Diet Coke
chipo.choice_description.value_counts().index[0]
```

```python
print('아이템 종류: ', len(chipo.item_name.value_counts()))
```
